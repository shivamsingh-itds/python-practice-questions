```python
# 1. User will input (3ages).Find the oldest one

age1 = int(input(f'Enter the age 1 : '))
age2 = int(input(f'Enter the age 2 : '))
age3 = int(input(f'Enter the age 3 : '))

if age1 > age2 and age1 > age3 :
    print(f"Age 1 is is oldest which is {age1}")
elif age2 > age1 and age2 > age3 :
    print(f"Age 2 is oldest which is {age2}")
else :
    print(f"Age 3 is oldest which is {age3}")
```

```python
# Write a program that will convert celsius value to fahrenheit

celsius = float(input(f"Enter the celsius : "))
fahrenheit = (celsius * 9/5) + 32
print(f"celsisus {celsius} is total " , fahrenheit , "fahrenheit")
```

```python
# User will input (2numbers).Write a program to swap the numbers

a = int(input("Number 1 :"))
b = int(input("Number 2 :"))
print(f"First number before swap is {a} and second number is {b}")
temp = b
b = a 
a = temp
print(f"first number after swap is {a} and second number after swap is {b}")
```

```python
# 5. Write a program that will reverse a four digit number.Also it checks whether the reverse is true.
num= int(input("Enter the 4 digit number  : "))
str1 = str(num)
var = str1[::-1]
print("Reverse is " ,var)
if num == int(var):
    print("it's palendrom ")
else :
    print("Not a palendrom")
```

```python
#7. Write a program that will tell whether the given year is a leap year or not.
year = int(input("Enter the year: "))

if (year%4 == 0 and year%100 != 0 ) or (year%400 ==0 ):
    print("It's is leap year")
else :
    print("It's not a leap year")
```
