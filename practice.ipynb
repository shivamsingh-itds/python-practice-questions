```python
# 1. User will input (3ages).Find the oldest one

age1 = int(input(f'Enter the age 1 : '))
age2 = int(input(f'Enter the age 2 : '))
age3 = int(input(f'Enter the age 3 : '))

if age1 > age2 and age1 > age3 :
    print(f"Age 1 is is oldest which is {age1}")
elif age2 > age1 and age2 > age3 :
    print(f"Age 2 is oldest which is {age2}")
else :
    print(f"Age 3 is oldest which is {age3}")
```

```python
# 8 Write a program to find the euclidean distance between two coordinates.
# D = ( x 2 − x 1 ) 2 + ( y 2 − y 1 ) 2

import math 

print("Enter the first coordinate (x1,x2)")
x1 = float(input("enter the cordinate x1 :"))
x2 = float(input("enter the cordinate x2 :" ))

print("Enter the second coordinate (y1,y2)")
y1 = float(input("enter the cordinate y1 :"))
y2 = float(input("enter the cordinate y2 :"))
ed = math.sqrt((x2-x1)**2+(y2-y1)**2)
print("The euclidean distance : " , ed)
```
